Q1.2. Answers last month

```python
# Using the stack overflow documentation, let´s take a look at the ANSWERS of last month

answers_lm = SITE.fetch('answers', fromdate=1577660400, todate=1580338800)

# We need to filter by a specific "title index" given it´s passing a lot of irrelevant data and we only want the
# questions. In this data it´s defined as "title"

items_2 = answers_lm['items']
items_2
```

```python
answers_list = []

for i in items_2:
    answers_list.append(i['answer_id'])
    
answers_list
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here

questions_most_voted = SITE.fetch('questions', fromdate=1577660400, todate=1580338800,
    min=5, tagged='python', sort='votes')

most_voted_question = questions_most_voted['items'][0]['title']

print("The most voted question is : ", most_voted_question)
```
